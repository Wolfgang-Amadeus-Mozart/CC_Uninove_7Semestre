## MINERACAO DE TEXTO - ARRUMANDO!

```python
df_textos = []
for i in df['DescricaoProblema']:
  df_textos.append(i)
```

```python
corpus = ' '.join(str(e) for e in df_textos)
corpus
```

```python
# Retorna a frequencia com que cada token aparece no texto 
from collections import Counter 
from nltk.tokenize import word_tokenize
dici = Counter(word_tokenize(corpus))
```

```python
caracteres = []
for i in sorted(dici, key = dici.get, reverse=True):
    caracteres.append(i)

caracteres
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(caracteres[:10], sorted(dici.values(), reverse=True)[:10])
```

```python
#como os primeiros colocados foram caracteres speciais, elimei-os
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(caracteres[4:14], sorted(dici.values(), reverse=True)[4:14])
```

```python
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords

#nltk.download() só poderá ser executado após instalar o nltk / pip install nltk
nltk.download('stopwords')

stops = stopwords.words('english')


mapa_cores = ListedColormap(['orange', 'green', 'red', 'magenta'])


# Criação da nuvem de palavras, com no máximo 100 palavras e utilizando as stop words
nuvem = WordCloud(background_color = 'white',
                  colormap = mapa_cores,
                  stopwords = stops,
                  max_words = 100)


# Criação e visualização da nuvem de palavras
nuvem.generate(corpus)

fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.imshow(nuvem)
```
