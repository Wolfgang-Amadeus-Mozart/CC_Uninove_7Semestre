```python
from sklearn.model_selection import train_test_split
from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
import numpy as np
```

```python
#Gerando a variavel tempo atendimento
df['DataArquivamento'] = pd.to_datetime(df['DataArquivamento'])
df['DataAbertura'] = pd.to_datetime(df['DataAbertura'])
df['Tempo_Atendimento'] = pd.to_datetime(df['DataArquivamento']) - pd.to_datetime(df['DataAbertura'])
```

```python
df['DataArquivamento'] = pd.to_numeric(df['DataArquivamento'])
df['DataAbertura'] = pd.to_numeric(df['DataAbertura'])
```

```python
#coletando somente a diferença do número de dias para um melhor aproveitamento dos dados
df['Tempo_Atendimento'] = df['Tempo_Atendimento'].dt.days 
```

# Feature engeneering

```python
#Feature Engeneering
df['SexoConsumidor'] = df['SexoConsumidor'].astype('category')
df['SexoConsumidor'] = df['SexoConsumidor'].cat.codes

df['Atendida'] = df['Atendida'].astype('category')
df['Atendida'] = df['Atendida'].cat.codes
```

```python
# Criar variáveis Dummie
df = pd.get_dummies(df, columns=['Regiao', 'UF','FaixaEtariaConsumidor'], drop_first=False)
```

```python
# Correlação entre as variáveis 
corr = df.corr()
fig2, ax = plt.subplots(figsize=(15,7))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            annot=True, cmap='YlGnBu', annot_kws={'size': 12}, fmt='.2f')
```

```python
#Seprando entre teste e treino
y = df['Tempo_Atendimento']
X = df.drop(columns=['Tempo_Atendimento', 'strRazaoSocial', 'strNomeFantasia', 'CNAEPrincipal', 'DescCNAEPrincipal',
                     'NumeroCNPJ', 'RadicalCNPJ', 'RazaoSocialRFB', 'DescricaoAssunto', 'DescricaoProblema','CEPConsumidor',
                    'DataArquivamento','DataAbertura'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, train_size=0.60, random_state=42)
```

```python
# Função Metrics para avaliação dos modelos
def metrics(y_test, y_pred):
    mae = round(mean_absolute_error(y_test, y_pred),1)
    mse = round(mean_squared_error(y_test, y_pred),1)
    rmse = round(np.sqrt(mse),1)
    r2 = round(r2_score(y_test, y_pred),2)
    print('MAE:{}, MSE: {}, RMSE: {}, R2: {}'.format(mae, mse, rmse, r2))
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100,
                                    criterion='squared_error', 
                                    max_depth=None, 
                                    min_samples_split=2, 
                                    min_samples_leaf=1, 
                                    min_weight_fraction_leaf=0.0, 
                                    max_features='auto', 
                                    max_leaf_nodes=None, 
                                    min_impurity_decrease=0.0, 
                                    bootstrap=True, 
                                    oob_score=False, 
                                    n_jobs=None, 
                                    random_state=None, 
                                    verbose=0, 
                                    warm_start=False,
                                    ccp_alpha=0.0, 
                                    max_samples=None)
```

```python
rf_regressor.fit(X_train, y_train)
y_pred = rf_regressor.predict(X_test)
```

```python
metrics(y_test, y_pred)
```

```python
rf_regressor.score(X_test, y_pred)
```

# Tunning de parâmetros

```python
#Removendo as dummies e verificando novamente

dummies = df.columns[25:]
y = df['Tempo_Atendimento']
X = df.drop(columns=['Tempo_Atendimento', 'strRazaoSocial', 'strNomeFantasia', 'CNAEPrincipal', 'DescCNAEPrincipal',
                     'NumeroCNPJ', 'RadicalCNPJ', 'RazaoSocialRFB', 'DescricaoAssunto', 'DescricaoProblema','CEPConsumidor',
                    'DataArquivamento','DataAbertura'])

X.drop(columns=dummies, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, train_size=0.60, random_state=42)
```

```python
#Acrescentando tunning de parametro
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

n_estimator = [1, 50, 100, 150, 200, 250, 300, 400]
min_samples_split = [2, 4, 8, 16]
min_samples_leaf = [1, 2, 4, 8]
max_features = ['auto', 'sqrt', 'log2', None]
max_depth = [2, 4, 6, 8, 10]
criterion = ['squared_error', 'absolute_error', 'poisson']

# Dicionário com os parametros
rf_params = {
    'n_estimators':n_estimator,
    'min_samples_split': min_samples_split,
    'min_samples_leaf':min_samples_leaf,
    'max_features':max_features,
    'max_depth':max_depth,
    'criterion':criterion
}

rf_tuned = RandomizedSearchCV(estimator=RandomForestRegressor(random_state=42), 
                              param_distributions=rf_params, # passagem dos parametros
                              cv=5, # validação cruzada 
                              n_iter=5, # numero de interações
                              scoring='recall', # metrica de avaliação 
                              n_jobs=-1, # uso de processadores
                              verbose=0) # log  
```
